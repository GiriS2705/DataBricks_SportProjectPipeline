# Git Operations for DataBricks_SportProjectPipeline

This notebook commits and pushes all files from consolidated_pipeline to GitHub.

**Repository:** https://github.com/GiriS2705/DataBricks_SportProjectPipeline.git

**Files to commit:** 8 notebooks across 3 folders
- 1_setup/ (3 notebooks)
- 2_dimension_data_processing/ (3 notebooks)
- 3_fact_data_processing/ (2 notebooks)

```python
import subprocess
import os

# Possible repository paths
possible_paths = [
    "/Workspace/Repos/giriwhatsapp27@gmail.com/DataBricks_SportProjectPipeline",
    "/Workspace/Users/giriwhatsapp27@gmail.com/DataBricks_SportProjectPipeline",
    "/repos/giriwhatsapp27@gmail.com/DataBricks_SportProjectPipeline"
]

repo_path = None

print("Searching for git repository...\n")

for path in possible_paths:
    print(f"Checking: {path}")
    # Check if directory exists and is a git repo
    if os.path.exists(path):
        result = subprocess.run(
            f"cd {path} && git status",
            shell=True,
            capture_output=True,
            text=True
        )
        if result.returncode == 0:
            repo_path = path
            print(f"✓ Found git repository at: {repo_path}\n")
            break
        else:
            print(f"  Not a git repo or not accessible")
    else:
        print(f"  Path does not exist")

if not repo_path:
    print("\n❌ Error: Could not find git repository!")
    print("\nPlease ensure DataBricks_SportProjectPipeline is set up as a Databricks Repo.")
    raise Exception("Git repository not found")
else:
    print(f"\n{'='*60}")
    print(f"Using repository path: {repo_path}")
    print(f"{'='*60}")
```

```python
print("\n" + "="*60)
print("STEP 1: Checking Git Status")
print("="*60 + "\n")

result = subprocess.run(
    f"cd {repo_path} && git status",
    shell=True,
    capture_output=True,
    text=True
)

if result.returncode == 0:
    print(result.stdout)
    
    # Check if there are changes to commit
    if "nothing to commit" in result.stdout:
        print("\n⚠️  No changes to commit. All files are already committed.")
    elif "Untracked files" in result.stdout or "Changes not staged" in result.stdout:
        print("\n✓ Found uncommitted changes to stage and commit.")
else:
    print(f"❌ Error checking git status:\n{result.stderr}")
    raise Exception("Git status failed")
```

```python
print("\n" + "="*60)
print("STEP 2: Staging All Files (git add .)")
print("="*60 + "\n")

result = subprocess.run(
    f"cd {repo_path} && git add .",
    shell=True,
    capture_output=True,
    text=True
)

if result.returncode == 0:
    print("✓ Successfully staged all files\n")
    
    # Show what was staged
    result = subprocess.run(
        f"cd {repo_path} && git status",
        shell=True,
        capture_output=True,
        text=True
    )
    print("Staged files:")
    print(result.stdout)
else:
    print(f"❌ Error staging files:\n{result.stderr}")
    raise Exception("Git add failed")
```

```python
print("\n" + "="*60)
print("STEP 3: Committing Changes")
print("="*60 + "\n")

commit_message = "Add consolidated pipeline - 8 notebooks across 3 processing stages (1_setup, 2_dimension_data_processing, 3_fact_data_processing)"

result = subprocess.run(
    f"cd {repo_path} && git commit -m '{commit_message}'",
    shell=True,
    capture_output=True,
    text=True
)

if result.returncode == 0:
    print("✓ Successfully committed changes\n")
    print(result.stdout)
elif "nothing to commit" in result.stdout:
    print("⚠️  Nothing to commit - all changes are already committed")
else:
    print(f"❌ Error committing changes:\n{result.stderr}")
    if result.stdout:
        print(result.stdout)
    raise Exception("Git commit failed")
```

```python
print("\n" + "="*60)
print("STEP 4: Pushing to GitHub")
print("="*60 + "\n")

print("Pushing to: https://github.com/GiriS2705/DataBricks_SportProjectPipeline.git\n")

result = subprocess.run(
    f"cd {repo_path} && git push",
    shell=True,
    capture_output=True,
    text=True
)

if result.returncode == 0:
    print("✓ Successfully pushed to GitHub!\n")
    print(result.stdout)
    if result.stderr:  # Git sometimes outputs to stderr even on success
        print(result.stderr)
else:
    print(f"❌ Error pushing to GitHub:\n{result.stderr}")
    if result.stdout:
        print(result.stdout)
    
    print("\n⚠️  Common issues:")
    print("  - Check if you have push permissions to the repository")
    print("  - Ensure your git credentials are configured in Databricks Repos")
    print("  - You may need to authenticate via Databricks Repos UI first")
    raise Exception("Git push failed")
```

```python
print("\n" + "="*60)
print("FINAL STATUS")
print("="*60 + "\n")

result = subprocess.run(
    f"cd {repo_path} && git status",
    shell=True,
    capture_output=True,
    text=True
)

if result.returncode == 0:
    print(result.stdout)
    
    # Get the latest commit info
    result = subprocess.run(
        f"cd {repo_path} && git log -1 --oneline",
        shell=True,
        capture_output=True,
        text=True
    )
    
    if result.returncode == 0:
        print(f"\n\n📝 Latest commit:")
        print(result.stdout)

print("\n" + "="*60)
print("✅ ALL OPERATIONS COMPLETED SUCCESSFULLY!")
print("="*60)
print("\n🎉 Your consolidated_pipeline code is now on GitHub!")
print("🔗 View at: https://github.com/GiriS2705/DataBricks_SportProjectPipeline")
```
